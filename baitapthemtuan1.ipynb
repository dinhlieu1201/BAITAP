```python
#Bài 8
ten = input("Ten chu ho: ")
chi_so_truoc = int(input("Chi so thang truoc: "))
chi_so_nay = int(input("Chi so thang nay: "))
so_kWh = chi_so_nay - chi_so_truoc
tien_dien = 0
if so_kWh > 0:
    if so_kWh <= 50:
        tien_dien = so_kWh * 1984
    elif so_kWh <= 100:
        tien_dien = 50 * 1984 + (so_kWh - 50) * 2050
    elif so_kWh <= 200:
        tien_dien = 50 * 1984 + 50 * 2050 + (so_kWh - 100) * 2380
    elif so_kWh <= 300:
        tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (so_kWh - 200) * 2998
    elif so_kWh <= 400:
        tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (so_kWh - 300) * 3350
    else:
        tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (so_kWh - 400) * 3460

tien_VAT = tien_dien * 0.08
tong_tien = int(tien_dien + tien_VAT)
print(f"Ho va ten: {ten}")
print(f"Tien phai tra la: {tong_tien}")
```
