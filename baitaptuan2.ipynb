```python
#10
import math
a,b=map(float,input('Nhập vào tâm đường tròn:').split())
r=float(input('Nhập bán kính đường tròn:'))
x,y=map(float,input('Nhập vào điểm A bất kì:').split())
OA=math.sqrt(((x-a)**2)+(y-b)**2)
if OA==r:
    print('Điểm A nằm trên đường tròn')
else:
    print('Điểm A không nằm trên đường tròn')
```
