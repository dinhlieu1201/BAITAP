```python
#W4A2
while True:
    n = int(input("Nhập một số: "))
    if n > 0:
        break

def la_so_nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

if la_so_nguyen_to(n):
    print(n, "là số nguyên tố")
else:
    print(n, "không phải là số nguyên tố")

        
```

```python
#W4A6
def la_so_nguyen_to(n):
    if n<2:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
while True:
    try:
        a, b = map(int, input().split())
        tong = 0
        for i in range(a, b + 1):
            if la_so_nguyen_to(i):
                tong += i
        print(tong)
        break
    except:
        break

    
```

```python
#W4A8
n = int(input())
steps = 0
while True:
    s=str(n)
    if s==s[::-1]:
        break
    rev=int(s[::-1])
    n=n+rev
    steps=steps+1
print(steps,n)
```

```python
#W4A9
n=int(input())
kq=[]
for i in range(1,int(n**0.5)+1):
    x=i*i
    s=str(x)
    if len(set(s))==len(s):
        kq.append(x)
if len(kq)==0:
    print('no number')
else:
    for x in kq:
        print(x,end=' ')
```

```python
#W4A10
n = int(input())
max_len = 0
best_x = 1
for x in range(1, n + 1):
    temp = x
    length = 1  
    while temp != 1:
        if temp % 2 == 0:
            temp //= 2
        else:
            temp = 3 * temp + 1
        length += 1
    if length > max_len:
        max_len = length
        best_x = x
print(best_x, max_len)
```

```python
#W4A13
a, b = map(int, input().split())
def tong_uoc(n):
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i
    return s
if tong_uoc(a) == b and tong_uoc(b) == a:
    print("true")
else:
    print("false")
```
