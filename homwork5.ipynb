```python
#W6A19
def maxvl(dsach):
    nhat=dsach[0]
    for i in range(1,len(dsach)):
        if dsach[i]>nhat:
            nhat=dsach[i]
    return i

dsach=list(map(int,input().split()))
print(maxvl(dsach))
```

```python
#W6A20
def findFirstIndex(arr, k):
    for i in range(len(arr)):
        if arr[i] == k:
            return i
    return -1

arr=list(map(int,input().split()))
k=int(input())
print(findFirstIndex(arr,k))
```

```python
#W6A23
def xoapt(arr,x):
    newarr=[]
    for i in arr:
        if i>=x:
            newarr.append(i)
    return newarr

arr=list(map(int,input().split()))
x=int(input())
print(xoapt(arr,x))
```

```python
#W6A24
def moveZeroToFront(arr):
    zeros = []
    others = []

    for x in arr:
        if x == 0:
            zeros.append(x)
        else:
            others.append(x)

    return zeros + others
```

```python
#W6A26
n = int(input())
a = list(map(int, input().split()))
tong = 0
for i in range(1, n - 1):
    if a[i] > a[i - 1] and a[i] > a[i + 1]:
        tong += a[i]

print(tong)
```

```python
#W6A28
n = int(input())
a = list(map(int, input().split()))

tang = True
giam = True

for i in range(n - 1):
    if a[i] >= a[i + 1]:
        tang = False
    if a[i] <= a[i + 1]:
        giam = False

if tang:
    print("DAY TANG")
elif giam:
    print("DAY GIAM")
else:
    print("VO DANH")
```
