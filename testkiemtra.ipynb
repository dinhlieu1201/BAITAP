```python
#W4A1 dùng for
n = int(input())
if n>1000:
    print('Vui lòng nhập n <=1000')
else:
    tong=0
    for i in range(1,n+1):
        tong +=i
    print(tong)
```

```python
#W4A2 dùng while
n = int(input())
if n>1000:
    print('Vui lòng nhập n <=1000')
else:
    tong = 0
    i =1 
    while i <= n:
        tong +=i
        i +=1
    print(tong)
```

```python
#test1
# Nhập số lượng phần tử
while True:
    try:
        n = int(input("Nhập số lượng phần tử n: "))
        if n > 0:
            break
        else:
            print("Số lượng phần tử phải lớn hơn 0!")
    except ValueError:
        print("Vui lòng nhập số nguyên!")

# Khởi tạo danh sách rỗng
arr = []

# Nhập các phần tử
for i in range(n):
    while True:
        try:
            value = int(input(f"Nhập phần tử thứ {i}: "))
            arr.append(value)
            break
        except ValueError:
            print("Vui lòng nhập số nguyên!")

# Tính tổng các phần tử ở vị trí chẵn (chỉ số chẵn)
total = 0
for i in range(0, n, 2):  # range(0, n, 2) tạo ra các chỉ số 0, 2, 4, ...
    total += arr[i]

# In kết quả
print(f"Dãy số vừa nhập: {arr}")
print(f"Tổng các phần tử ở vị trí chẵn (chỉ số chẵn) là: {total}")
```

```python
#test2
numbers = list(map(int, input().split()))
if len(numbers) == 0:
    print("Không có số nào được nhập!")
else:
    # Tìm giá trị lớn nhất
    max_value = numbers[0]
    for num in numbers:
        if num > max_value:
            max_value = num
    print(max_value)
```

```python
#test3 làm tròn xuống
m, n = map(int, input().split())
ket_qua = m // n
print(ket_qua)
```
