```python
#Bài 4
h = int(input())
for i in range(1, h):
    print(" " * (h - i), end="")
    print("*", end="")
    if i > 1:
        print(" " * (2 * (i - 1) - 1), end="")
        print("*", end="")
    print()  
if h > 0:
    print("*" * (2 * h - 1))
```

```python
#Bài 14
x=int(input())
if x==-1:
    print('chưa có số nào được nhập')
else:
    maxvl=x
    minvl=x
    while True:
        x = int(input())
        if x == -1:
            break
        if x > maxvl:
            maxvl = x
        if x < minvl:
            minvl = x
    print("Số lớn nhất là:", maxvl)
    print("Số nhỏ nhất là:", minvl)
```

```python
#Bài 15
n = int(input())
dem = 0
tam = n

while tam > 0:
    dem += 1
    tam //= 10

print("Số chữ số của", n, "là:", dem)
```

```python
#Bài 16
n=int(input())
chan=0
le=0
while n>0:
    chuso=n%10
    if chuso%2==0:
        chan=chan+1
    else:
        le=le+1
    n=n//10
print('có',chan,'chữ số chẵn và',le,'chữ số lẻ')
```

```python
#Bài 17
n = int(input())
tong = 0
while n > 0:
    tong += n % 10
    n //= 10

print("Tổng các chữ số là:", tong)
```

```python
#Bài 18
n = int(input())
tam = n
while tam> 1:
    if tam % 3 != 0:
        print(n, "không phải là số dạng 3^k")
        break
    tam //= 3
else:
    print(n, "là số dạng 3^k")
```

```python
#Bài 19
a = int(input())
f1 = 1
f2 = 1
while f2 <= a:
    f1, f2 = f2, f1 + f2
print("Số Fibonacci lớn nhất không vượt quá a là:", f1)
```

```python
#Bài 20
n = int(input())
dem= 0
while n > 0:
    chuso = n % 10
    if chuso in (2, 3, 5, 7):
        dem += 1
    n //= 10

print("Số chữ số nguyên tố là:", dem)
```
