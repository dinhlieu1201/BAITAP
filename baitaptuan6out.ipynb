```python
#W6A1
arr=list(map(int,input().split()))
d=[]
for i in arr:
    if i not in d:
        d.append(i)
print(d)
```

```python
#W6A2
arr=list(map(int,input().split()))
tong=0
d=[]
for i in arr:
    tong+=i
    d.append(tong)
print(d)

```

```python
#W6A3
arr=tuple(map(int,input().split()))
k=int(input())
n=len(arr)
res=[]
for i in range(n):
    if i==k:
        res=arr[k:]+arr[:k]
print(res)
```

```python
#W6A4
dic=input().split()
d={}
for i in dic:
    key,value=i.split(':')
    if key not in d:
        d[key]=[value]
    else:
        d[key].append(value)
   
print(d)
```

```python
#W6A5
arr=list(map(int,input().split()))
d={'positives':0,'negatives':0,'zeros':0}
for i in arr:
    if i>0:
        d['positives']+=1
    if i<0:
        d['negatives']+=1
    if i==0:
        d['zeros']+=1
print(d)
```
