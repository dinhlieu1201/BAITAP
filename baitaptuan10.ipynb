```python
#W9A5
def is_subarray(list1, list2):
    n = len(list1)
    m = len(list2)

    if n > m:
        return False

    for i in range(m - n + 1):
        ok = True
        for j in range(n):
            if list2[i + j] != list1[j]:
                ok = False
                break
        if ok:
            return True

    return False
a = list(map(int, input().split()))
b = list(map(int, input().split()))

if is_subarray(a, b):
    print('YES')
else:
    print('NO')
```

```python
#W9A6
n = int(input())
a = list(map(int, input().split()))

i = n - 2
while i >= 0 and a[i] > a[i + 1]:
    i -= 1
if i < 0:
    a.reverse()
else:
    j = n - 1
    while a[j] < a[i]:
        j -= 1
    a[i], a[j] = a[j], a[i]
    a[i + 1:] = reversed(a[i + 1:])
for x in a:
    print(x, end=" ")
```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)

    for i in range(n):
        if flowerbed[i] == 0:
            left_ok = (i == 0) or (flowerbed[i - 1] == 0)
            right_ok = (i == n - 1) or (flowerbed[i + 1] == 0)

            if left_ok and right_ok:
                flowerbed[i] = 1
                k -= 1
                if k == 0:
                    return True

    return k <= 0
flowerbed=list(map(int,input().split()))
k=int(input())
print(canPlaceFlowers(flowerbed,k))
```

```python
#W9A8
def is_path_crossing(moves):
    x, y = 0, 0
    visited = set()
    visited.add((0, 0))   # vị trí ban đầu

    for m in moves:
        if m == 'R':
            x += 1
        elif m == 'L':
            x -= 1
        elif m == 'U':
            y += 1
        elif m == 'D':
            y -= 1

        if (x, y) in visited:
            return True

        visited.add((x, y))

    return False
```

```python
#W9A9
def Median(nums):
    a, b, c, d, e = nums
    if a > b:
        a, b = b, a
    if c > d:
        c, d = d, c
    if a > c:
        a, c = c, a
        b, d = d, b
    if b > d:
        b, d = d, b
    if e < b:
        if e < c:
            return c
        else:
            return e
    else:
        if e < d:
            return e
        else:
            return b
nums=list(map(int,input().split()))
print(Median(nums))
```
