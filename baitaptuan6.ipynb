```python
#W6A4
dict=input().split()
d={}
for i in dict:
    key,value = i.split(':')
    value=int(value)
    if key not in d:
        d[key]=[value]
    else:
        d[key].append(value)
print(d)
```

```python
#W6A5
nums = list(map(int, input().split()))
res={'positives':0,'negatives':0,'zeros':0}
for i in nums:
    if i>0:
        res['positives']+=1
    elif i<0:
        res['negatives']+=1
    else:
        res['zeros']+=1
print(res)
```

```python
#W6A9
dt1= input().split()
d1={}
for x in dt1:
    key,value=x.split(':')
    d1[key]=int(value)

dt2=input().split()
d2={}
for x in dt2:
    key,value=x.split(':')
    d2[key]=int(value)

res={}
for key in d1:
    res[key]=d1[key]
for key in d2:
    if key in res:
        res[key]+=d2[key]
    else:
        res[key]=d2[key]
res_sorted = dict(sorted(res.items()))
print(res_sorted)

```

```python
#W6A10
dsach=list(map(int,input().split()))
k=int(input())
res=[]
n=len(dsach)
for i in range(n):
    for j in range(i+1,n):
        if dsach[i]+dsach[j]==k:
            res.append(dsach[i],dsach[j])
res.sort(key=lambda x:x[0])
print(res)
```

```python
#W6A12
n=list(map(int,input().split()))
count={}
for i in n:
    if i not in count:
        count[i]=1
    else:
        count[i]+=1
nhieunhat=max(count.values())

ans = None
for x in count:
    if count[x] == nhieunhat:
        if ans is None or x < ans:
            ans = x

print(ans)
```
