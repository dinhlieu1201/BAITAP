```python
#Bài 3
a,b,c=map(int,input().split())
if a+b>c and a+b>b and b+c>a:
    if a==b==c:
        print('Đây là tam giác đều')
    elif a==b and a==c or a==b and b==c or b==c and a==c:
        print('Đây là tam giác cân')
    elif (a**2==b**2+c**2) or (b**2==a**2+c**2) or (c**2==a**2+b**2):
        print('Đây là tam giác vuông')
    else:
        print('Đây là tam giác thường')
    p=(a+b+c)/2
    dientich=(p*(p-a)*(p-b)*(p-c))**0.5
    chuvi=a+b+c
    print('Chu vi hinh tam giac la:',chuvi)
    print('Dien tich hinh tam giac la:',dientich)
else:
    print('Không phải 3 cạnh của tam giác')

```
