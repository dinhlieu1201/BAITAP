```python
#W7A16
arr=map(int,input().split())
removed=0
checked=[]
for x in arr:
    if x in checked:
        continue
    dem=0
    for i in arr:
        if i==x:
            dem+=1
    if dem>x:
        removed+=dem-x
    elif dem<x:
        removed+=dem
    checked.append(x)
if removed==0:
    print('YES')
else:
    print('NO',removed)

```

```python
#W7A22
n = int(input())

d = []      # danh sách tên riêng
dem = []    # số lần xuất hiện

for i in range(n):
    full = input().strip()
    part = full.split()
    ten = part[-1]      

    if ten in d:
        idx = d.index(ten)
        dem[idx] += 1
    else:
        d.append(ten)
        dem.append(1)

# tìm tên xuất hiện nhiều nhất
result = d[0]
max_count = dem[0]

for i in range(len(d)):
    if dem[i] > max_count:
        max_count = dem[i]
        result = d[i]
    elif dem[i] == max_count:
        if d[i] < result:
            result = d[i]

print(result)



```

```python
#W7A23
s=input()
words=s.split()
dem=0
for i in words:
    if len(i)<2:
        continue
    if not('A'<=i[0]<='Z'):
        continue
    ok=True
    for ch in i[1:]:
        if not ('a'<=ch<='z'):
            ok=False
            break
    if ok:
        dem+=1
print(dem)
```

```python
#W7A27
s = input().strip()

stack = []
valid = True

for ch in s:
    if ch == '(' or ch == '{' or ch == '[':
        stack.append(ch)
    else:
        if len(stack) == 0:
            valid = False
            break

        top = stack.pop()

        if ch == ')' and top != '(':
            valid = False
            break
        if ch == '}' and top != '{':
            valid = False
            break
        if ch == ']' and top != '[':
            valid = False
            break

if valid and len(stack) == 0:
    print("yes")
else:
    print("no")
```
