```python
#W7A16
arr=map(int,input().split())
removed=0
checked=[]
for x in arr:
    if x in checked:
        continue
    dem=0
    for i in arr:
        if i==x:
            dem+=1
    if dem>x:
        removed+=dem-x
    elif dem<x:
        removed+=dem
    checked.append(x)
if removed==0:
    print('YES')
else:
    print('NO',removed)

```
