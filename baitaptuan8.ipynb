```python
#W7A16
arr=map(int,input().split())
removed=0
checked=[]
for x in arr:
    if x in checked:
        continue
    dem=0
    for i in arr:
        if i==x:
            dem+=1
    if dem>x:
        removed+=dem-x
    elif dem<x:
        removed+=dem
    checked.append(x)
if removed==0:
    print('YES')
else:
    print('NO',removed)

```

```python
#W7A18
arr=list(map(int,input().split()))
chan=[]
le=[]
for i in arr:
    if i%2==0:
        chan.append(i)
    else:
        le.append(i)
chan.sort()
le.sort(reverse=True)
for x in chan+le:
    print(x,end=' ')
```

```python
#W7A19
#2 mảng tăng dần
a=list(map(int,input().split()))
b=list(map(int,input().split()))
i=0
j=0
d=[]
while i<len(a) and j<len(b):
    if a[i]<=b[j]:
        d.append(a[i])
        i+=1
    else:
        d.append(b[j])
        j+=1
while i<len(a):
    d.append(a[i])
    i+=1
while j<len(b):
    d.append(b[j])
    j+=1
for x in d: 
    print(x, end=" ")
```

```python
#W7A20
arr=map(int,input().split())
k=int(input())
tong=0
for i in arr:
    if len(str(i))==k:
        tong+=i
print(tong)
```

```python
#W7A20 CÁCH 2 KHÔNG DÙNG len(str(i))
arr=map(int,input().split())
k=int(input())
tong=0
for i in arr:
    dem=0
    tam=i
    while tam>0:
        dem+=1
        tam=tam//10
    if dem==k:
        tong+=i
print(tong)
```

```python
#W7A21
N,T=map(int,input().split())
days=[31,29,31,30,31,30,31,30,31,30,31,30]
if T<1 or T > 12:
    print(-1)
else:
    if N < 1 or N > days[T - 1]:
        print(-1)
    else:
        total = 0
        for i in range(T - 1):
            total += days[i]
        total += N
        print(total)
```
