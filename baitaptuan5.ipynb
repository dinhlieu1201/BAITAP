```python
#W5A1
def Max_Of_Two(a, b) -> int:
    maxvl=a
    if b>maxvl:
        maxvl=b
    return maxvl
a,b=map(int,input().split())
print(Max_Of_Two(a,b))
```

```python
#W5A2
def swap(a, b):
    tam=a
    a=b
    b=tam
    return a,b
a,b=map(int,input().split())
print(swap(a,b))
```

```python
#W5A2 cách 2
from typing import Tuple
def swap(a: int, b: int) -> Tuple[int, int]:
    return b, a
a,b=map(int,input().split())
print(swap(a,b))
```

```python
#W5A3
def ktra(n):
    if n<2:
        return False
    for i in range(2,n):
        if n%i==0:
            return False
    return True
n=int(input())
print(ktra(n))
```

```python
#W5A4
def hoanhao(n:int)->bool:
    if n<=1:
        return False
    s=0
    for i in range(1,n):
        if n%i==0:
            s=s+i
        if s==n:
            return True
    return False
n=int(input())
print(hoanhao(n))
```

```python
#W5A5
def vitri(a: list[int], k: int) -> int:
    for i in range(len(a)):
        if a[i] == k:
            return i
    return -1
day=list(map(int,input().split()))
a=int(input())
print(vitri(day,a))
```

```python
#W5A6
def giaithua(n):
    if n<0:
        return 'Vui lòng nhập lại số nguyên dương'
    giaithua=1
    for i in range(1,n+1):
        giaithua*=i
    return giaithua

n=int(input())
print(giaithua(n))
```
